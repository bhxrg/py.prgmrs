```python
#replace null with one value
dsn=pd.read_csv("C:\\Users\\bhargavi\\python\\data_cleaning.csv")
dsn["Calories"].fillna(130,inplace=True)
pd.isnull(dsn).sum()
```

```python
#replace null with mean
dsn["MaxPulse"].fillna(dsn["MaxPulse"].mean(),inplace=True)
pd.isnull(dsn).sum()
```

```python
#replace with median
dsn["Pulse"].fillna(dsn["Pulse"].median(),inplace=True)
pd.isnull(dsn).sum()
```

```python
#replace with max /min
dsn=pd.read_csv("C:\\Users\\bhargavi\\python\\data_cleaning.csv")
dsn["Pulse"].fillna(dsn["Pulse"].max(),inplace=True)
pd.isnull(dsn).sum()
```
